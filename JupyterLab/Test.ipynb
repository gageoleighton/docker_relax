# Widget at plotting in matplotlib

* [Widget at plotting in matplotlib](http://nbviewer.jupyter.org/github/jenshnielsen/netcdf-lesson/blob/master/Interactive%20plotting%20with%20nbagg.ipynb)


```python
from ipywidgets import interact, fixed, fixed
import ipywidgets as widgets

import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg
```

```python
def pltsin(ax, f=3):
    x = np.linspace(0,1,100)
    y = np.sin(2*np.pi*x*f)
    if len(ax.lines):
        ax.lines[0].set_xdata(x)
        ax.lines[0].set_ydata(y)
    else:
        ax.plot(x,y)
    fig.canvas.draw()
```

```python
fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(-1,1)
pltsin(ax)
```

```python
interact(pltsin, ax=fixed(ax), f=(1, 10, 0.1))
```
